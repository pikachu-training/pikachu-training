Merging and Joining

```python
contacts = pd.DataFrame(
    {
     'Name' :['Alice','Charlie'] ,
     'Tel'  :['+91 999 999 99 99', '+91 888 888 88 88'] 
    }
)
```

```python
contacts
```

```python
exam_results.merge(contacts, left_on='Name', right_on='Name')  # defaults to inner join
```

```python
exam_results.merge(contacts, left_on='Name', right_on ='Name', how ='outer')
```
